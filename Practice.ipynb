```python
def right_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        print()

def inverted_right_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n, 0, -1):
        for j in range(i):
            print(alphabet[j], end=' ')
        print()

def left_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        print(' ' * (n - i - 1), end='')
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        print()

def inverted_left_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        print(' ' * i, end='')
        for j in range(n - i):
            print(alphabet[j], end=' ')
        print()

def pyramid(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        print(' ' * (n - i - 1), end='')
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        for j in range(i):
            print(alphabet[i - j - 1], end=' ')
        print()

def inverted_pyramid(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n, 0, -1):
        print(' ' * (n - i), end='')
        for j in range(i):
            print(alphabet[j], end=' ')
        for j in range(i - 1):
            print(alphabet[i - j - 2], end=' ')
        print()

def diamond(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        print(' ' * (n - i - 1), end='')
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        for j in range(i):
            print(alphabet[i - j - 1], end=' ')
        print()
    for i in range(n - 2, -1, -1):
        print(' ' * (n - i - 1), end='')
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        for j in range(i):
            print(alphabet[i - j - 1], end=' ')
        print()

def right_pascals_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        print()
    for i in range(n - 1):
        for j in range(n - i - 1):
            print(alphabet[j], end=' ')
        print()

def left_pascals_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        print(' ' * (n - i - 1), end='')
        for j in range(i + 1):
            print(alphabet[j], end=' ')
        print()
    for i in range(n - 1):
        print(' ' * (i + 1), end='')
        for j in range(n - i - 1):
            print(alphabet[j], end=' ')
        print()

def sandglass(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n, 0, -1):
        print(' ' * (n - i), end='')
        for j in range(i):
            print(alphabet[j], end=' ')
        print()
    for i in range(1, n + 1):
        print(' ' * (n - i), end='')
        for j in range(i):
            print(alphabet[j], end=' ')
        print()

def reverse_right_triangle(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        for j in range(n - i):
            print(alphabet[j], end=' ')
        print()

def hollow_pyramid(n):
    alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for i in range(n):
        print(' ' * (n - i - 1), end='')
        for j in range(i + 1):
            if j == 0 or j == i:
                print(alphabet[j], end=' ')
            else:
                print(' ', end=' ')
        print()

n = 5
right_triangle(n)
print()
inverted_right_triangle(n)
print()
left_triangle(n)
print()
inverted_left_triangle(n)
print()
pyramid(n)
print()
inverted_pyramid(n)
print()
diamond(n)
print()
right_pascals_triangle(n)
print()
left_pascals_triangle(n)
print()
sandglass(n)
print()
reverse_right_triangle(n)
print()
hollow_pyramid(n)
```

```python
def fibonacci_sequence(n):
    sequence = []
    a, b = 0, 1
    while len(sequence) < n:
        sequence.append(a)
        a, b = b, a + b
    return sequence

# Number of terms to generate
n_terms = int(input("Enter the number of terms: "))

# Generate and print the Fibonacci sequence
fib_sequence = fibonacci_sequence(n_terms)
print("Fibonacci sequence:")
for num in fib_sequence:
    print(num, end=' ')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_spur_gear(num_teeth, pitch_circle_radius):
    fig, ax = plt.subplots()

    # Parameters for the gear
    pitch_radius = pitch_circle_radius
    tooth_angle = 2 * np.pi / num_teeth
    addendum = 0.1 * pitch_radius  # Addendum (length of teeth)
    dedendum = 0.15 * pitch_radius  # Dedendum (depth of teeth)

    # Draw gear teeth
    for i in range(num_teeth):
        theta = i * tooth_angle
        x = [pitch_radius * np.cos(theta), (pitch_radius + addendum) * np.cos(theta)]
        y = [pitch_radius * np.sin(theta), (pitch_radius + addendum) * np.sin(theta)]
        ax.plot(x, y, color='black')

    # Draw pitch circle
    pitch_circle = plt.Circle((0, 0), pitch_radius, fill=False, edgecolor='blue')
    ax.add_artist(pitch_circle)

    # Draw outer circle (just for visual reference)
    outer_circle = plt.Circle((0, 0), pitch_radius + addendum + dedendum, fill=False, linestyle='--', edgecolor='gray')
    ax.add_artist(outer_circle)

    # Set equal scaling and labels
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title(f'Spur Gear with {num_teeth} Teeth')

    plt.grid(True)
    plt.show()

# Define parameters for the gear
num_teeth = 68
pitch_circle_radius = 5  # Adjust this radius as needed for visual representation

# Draw the spur gear
draw_spur_gear(num_teeth, pitch_circle_radius)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given data
P_hp = 125  # Power transmitted in horsepower
N = 250  # Shaft speed in rpm
D_inches = 2  # Shaft diameter in inches
tau_design = 19000  # Design stress for shear in psi
sigma_design = 37000  # Design stress for bearing in psi

# Convert shaft diameter from inches to meters
D = D_inches * 0.0254  # Convert inches to meters

# Convert power from horsepower to watts
P_watts = P_hp * 745.7  # 1 horsepower = 745.7 watts

# Calculate torque (T) in N*m
T = P_watts * 60 / (2 * np.pi * N)  # P = 2 * pi * N * T / 60

# Initialize dimensions of the key
l = 0.02  # Initial length of the key (meters)
h = 0.01  # Initial height of the key (meters)

# Function to calculate shear stress (tau) on the key
def calculate_tau(l, h):
    return T / (tau_design * l * h / 2)

# Function to calculate bearing stress (sigma) on the key
def calculate_sigma(l, h):
    return 2 * T / (D * h)

# Iterate to find appropriate dimensions of the key
while calculate_tau(l, h) > 1 or calculate_sigma(l, h) > 1:
    if calculate_tau(l, h) > 1:
        l += 0.001  # Increase length incrementally
    if calculate_sigma(l, h) > 1:
        h += 0.001  # Increase height incrementally

# Print the final dimensions of the key
print(f"Dimensions of the key - Length (l): {l:.3f} m, Height (h): {h:.3f} m")

# Visualize key dimensions
# 2D diagram
plt.figure(figsize=(10, 6))
plt.plot([0, l], [0, 0], 'k-', linewidth=4)  # Bottom edge of the key
plt.plot([0, l], [h, h], 'k-', linewidth=4)  # Top edge of the key
plt.plot([0, 0], [0, h], 'k-', linewidth=2)  # Left edge of the key
plt.plot([l, l], [0, h], 'k-', linewidth=2)  # Right edge of the key
plt.xlim(-0.1, l + 0.1)
plt.ylim(-0.1, h + 0.1)
plt.xlabel('Length (m)')
plt.ylabel('Height (m)')
plt.title('2D Diagram of Flat Key')
plt.grid(True)
plt.savefig('flat_key_2d_diagram.png')  # Save the diagram
plt.show()

# 3D diagram
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, l], [0, 0], [0, 0], color='black', linewidth=4)  # Bottom surface of the key
ax.plot([0, l], [h, h], [0, 0], color='black', linewidth=4)  # Top surface of the key
ax.plot([0, 0], [0, h], [0, 0], color='black', linewidth=2)  # Left edge of the key
ax.plot([l, l], [0, h], [0, 0], color='black', linewidth=2)  # Right edge of the key
ax.set_xlabel('Length (m)')
ax.set_ylabel('Height (m)')
ax.set_zlabel('Width (m)')
ax.set_title('3D Diagram of Flat Key')
plt.savefig('flat_key_3d_diagram.png')  # Save the diagram
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def plot_hydraulic_press():
    # Create data for the base plate of the press
    base_plate_x = np.array([0, 10, 10, 0, 0])
    base_plate_y = np.array([0, 0, 2, 2, 0])
    base_plate_z = np.array([0, 0, 0, 0, 0])

    # Create data for the moving part (piston and attached plate)
    piston_x = np.array([3, 7, 7, 3, 3])
    piston_y = np.array([0, 0, 0.5, 0.5, 0])
    piston_z = np.array([0, 0, 0, 0, 0])

    # Create a figure for 2D plot
    fig, ax = plt.subplots()
    ax.plot(base_plate_x, base_plate_y, 'b-', linewidth=2, label='Base Plate')
    ax.plot(piston_x, piston_y, 'r-', linewidth=2, label='Piston')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Hydraulic Press 2D View')
    ax.legend()
    ax.grid(True)

    plt.show()

    # Create a 3D plot for perspective view
    fig_3d = plt.figure()
    ax3d = fig_3d.add_subplot(111, projection='3d')

    ax3d.plot(base_plate_x, base_plate_y, base_plate_z, 'b-', linewidth=2, label='Base Plate')
    ax3d.plot(piston_x, piston_y, piston_z, 'r-', linewidth=2, label='Piston')

    ax3d.set_xlabel('X')
    ax3d.set_ylabel('Y')
    ax3d.set_zlabel('Z')
    ax3d.set_title('Hydraulic Press 3D View')
    ax3d.legend()
    ax3d.grid(True)

    plt.show()

# Plot the hydraulic press simulation
plot_hydraulic_press()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def hydraulic_press_simulation(force, thickness):
    # Generate data for the squeezed material
    x = np.linspace(0, 1, 100)  # Sample points along x-axis
    y = np.linspace(0, thickness, 100)  # Sample points along y-axis
    X, Y = np.meshgrid(x, y)
    Z = force * (1 - Y/thickness)  # Simple linear compression model

    # Plot 2D image
    plt.figure(figsize=(10, 6))
    plt.contourf(X, Y, Z, cmap='viridis')
    plt.colorbar(label='Pressure')
    plt.xlabel('Width')
    plt.ylabel('Thickness')
    plt.title('Hydraulic Press 2D Simulation')
    plt.grid(True)
    plt.gca().invert_yaxis()  # Invert y-axis for better visualization
    plt.savefig('hydraulic_press_2d.png')  # Save the image
    plt.show()

    # Plot 3D image
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('Width')
    ax.set_ylabel('Thickness')
    ax.set_zlabel('Pressure')
    ax.set_title('Hydraulic Press 3D Simulation')
    plt.savefig('hydraulic_press_3d.png')  # Save the image
    plt.show()

# Parameters for the hydraulic press
force_applied = 1000  # Arbitrary force applied (adjust as needed)
material_thickness = 10  # Thickness of the material being squeezed

# Run the simulation and plot images
hydraulic_press_simulation(force_applied, material_thickness)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def hydraulic_press_punching_simulation(punch_force, material_thickness, hole_radius):
    # Generate data for the punched material
    x = np.linspace(-1, 1, 100)  # Sample points along x-axis
    y = np.linspace(-1, 1, 100)  # Sample points along y-axis
    X, Y = np.meshgrid(x, y)

    # Calculate distance from the center (assumed punch location)
    distance_from_center = np.sqrt(X**2 + Y**2)

    # Determine the pressure distribution (simple model)
    pressure = np.zeros_like(distance_from_center)
    pressure[distance_from_center <= hole_radius] = punch_force

    # Plot 2D image
    plt.figure(figsize=(10, 6))
    plt.contourf(X, Y, pressure, cmap='viridis')
    plt.colorbar(label='Pressure')
    plt.xlabel('Width')
    plt.ylabel('Height')
    plt.title('Hydraulic Press Punching 2D Simulation')
    plt.grid(True)
    plt.savefig('hydraulic_press_punching_2d.png')  # Save the image
    plt.show()

    # Plot 3D image
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, pressure, cmap='viridis')
    ax.set_xlabel('Width')
    ax.set_ylabel('Height')
    ax.set_zlabel('Pressure')
    ax.set_title('Hydraulic Press Punching 3D Simulation')
    plt.savefig('hydraulic_press_punching_3d.png')  # Save the image
    plt.show()

# Parameters for the hydraulic press punching simulation
punch_force_applied = 5000  # Force applied by the punch (adjust as needed)
material_thickness = 2  # Thickness of the material being punched
hole_radius = 0.5  # Radius of the punched hole

# Run the simulation and plot images
hydraulic_press_punching_simulation(punch_force_applied, material_thickness, hole_radius)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gas_welding_simulation(heat_intensity, material_thickness):
    # Parameters for the welding torch and material
    torch_position = (0, 0)  # Position of the welding torch
    torch_radius = 0.2  # Radius of the welding torch
    material_length = 4  # Length of the material being welded

    # Generate data for the material
    x = np.linspace(-1, material_length, 200)
    y = np.linspace(-1, 1, 100)
    X, Y = np.meshgrid(x, y)

    # Calculate temperature distribution (simple model)
    temperature = heat_intensity * np.exp(-((X - torch_position[0])**2 + (Y - torch_position[1])**2) / torch_radius**2)

    # Plot 2D image
    plt.figure(figsize=(10, 6))
    plt.contourf(X, Y, temperature, cmap='hot', levels=20)
    plt.colorbar(label='Temperature')
    plt.scatter([torch_position[0]], [torch_position[1]], color='blue', marker='o', label='Welding Torch')
    plt.xlabel('Length')
    plt.ylabel('Width')
    plt.title('Gas Welding Simulation')
    plt.legend()
    plt.grid(True)
    plt.savefig('gas_welding_simulation.png')  # Save the image
    plt.show()

# Parameters for the gas welding simulation
heat_intensity = 100  # Intensity of heat applied by the welding torch
material_thickness = 1  # Thickness of the material being welded

# Run the simulation and plot image
gas_welding_simulation(heat_intensity, material_thickness)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Plot the synthetic data
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Data")
plt.show()

# Initialize parameters
m = np.random.randn(1)
b = np.random.randn(1)

# Set learning rate and number of iterations
learning_rate = 0.1
iterations = 1000

# Store the values of m and b for plotting
m_history = []
b_history = []
loss_history = []

# Define loss function and gradient calculation
def compute_loss(X, y, m, b):
    n = len(y)
    predictions = m * X + b
    loss = (1/n) * np.sum((y - predictions) ** 2)
    return loss

def compute_gradients(X, y, m, b):
    n = len(y)
    predictions = m * X + b
    dm = -(2/n) * np.sum(X * (y - predictions))
    db = -(2/n) * np.sum(y - predictions)
    return dm, db

# Gradient Descent
for i in range(iterations):
    m_history.append(m[0])
    b_history.append(b[0])
    loss = compute_loss(X, y, m, b)
    loss_history.append(loss)

    dm, db = compute_gradients(X, y, m, b)

    m -= learning_rate * dm
    b -= learning_rate * db

    if i % 100 == 0:
        print(f"Iteration {i}: m = {m[0]}, b = {b[0]}, loss = {loss}")

# Plot loss over iterations
plt.plot(loss_history)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss Over Iterations")
plt.show()

# Plot the fitted line
plt.scatter(X, y, label="Data points")
plt.plot(X, m * X + b, color="red", label="Fitted line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```

Explanation
Batch Gradient Descent: The parameters are updated using the entire dataset, and the 3D plot shows a smooth path towards the minimum loss.
Stochastic Gradient Descent (SGD): The parameters are updated for each training example, and the 3D plot shows a more erratic path due to the high variance in the updates.
Mini-Batch Gradient Descent: The parameters are updated using small batches of data, providing a balance between the smooth path of Batch Gradient Descent and the high variance of SGD.
By plotting the evolution of the parameters

m and
𝑏
b along with the loss in 3D, we can visualize how each gradient descent method navigates the loss surface to find the optimal parameters
